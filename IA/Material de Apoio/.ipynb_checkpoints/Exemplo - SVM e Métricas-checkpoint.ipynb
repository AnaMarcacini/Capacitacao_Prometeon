```python
import pandas as pd

DataFrame = pd.read_csv("Database/WineDataset.csv")
DataFrame.head(10)
```

```python
for i in range(0, len(DataFrame)):
    
    if DataFrame.at[i, "color"] == "red":
        DataFrame.at[i, "color"] = 0
    
    elif DataFrame.at[i, "color"] == "white":
        DataFrame.at[i, "color"] = 1

DataFrame.head(10)
```

```python
DataFrame["color"].value_counts()
```

```python
DataFrame.tail()
```

```python
X = []
Y = []

for i in range(0, len(DataFrame)):
    Atributos = []
    
    for atributo in range(1,14):
        Atributos.append(DataFrame.iat[i,atributo]) # indice iat indice at (em)
    
    X.append(Atributos)
    Y.append(DataFrame.iat[i,14])

print("X (ENTRADA): ", X[0])
print("Y (SAÍDA): ", Y[0])
```

```python
from sklearn.model_selection import train_test_split

X_Treinamento, X_Teste, Y_Treinamento, Y_Teste = train_test_split(X, Y, test_size=0.3)

print("1ª Amostra Treino X (Entrada): ", X_Treinamento[0])
print("1ª Amostra Treino Y (Saída): ", Y_Treinamento[0])
print("1ª Amostra Teste X (Entrada): ", X_Teste[0])
print("1ª Amostra Teste Y (Saída): ", Y_Teste[0])
```

```python
from sklearn import svm
from sklearn.metrics import *

#criando modelos
ClassificadorSVM_Linear = svm.SVC(kernel="linear")
ClassificadorSVM_Polinomial = svm.SVC(kernel="poly")
ClassificadorSVM_RBF = svm.SVC(kernel="rbf")
ClassificadorSVM_Sigmoid = svm.SVC(kernel="sigmoid")
# treinando modelo
ClassificadorSVM_Linear.fit(X_Treinamento, Y_Treinamento)
ClassificadorSVM_Polinomial.fit(X_Treinamento, Y_Treinamento)
ClassificadorSVM_RBF.fit(X_Treinamento, Y_Treinamento)
ClassificadorSVM_Sigmoid.fit(X_Treinamento, Y_Treinamento)
# prevendo os valores com os dados de teste
LinearPredict = ClassificadorSVM_Linear.predict( X_Teste )
PolinomialPredict = ClassificadorSVM_Polinomial.predict( X_Teste )
RBFPredict = ClassificadorSVM_RBF.predict( X_Teste )
SigmoidPredict = ClassificadorSVM_Sigmoid.predict( X_Teste )
# Calculo da Acuracia
AcuraciaLinear = accuracy_score(Y_Teste, LinearPredict)
AcuraciaPolinomial = accuracy_score(Y_Teste, PolinomialPredict)
AcuraciaRBF = accuracy_score(Y_Teste, RBFPredict)
AcuraciaSigmoid = accuracy_score(Y_Teste, SigmoidPredict)
# Calculo da Precisão
PrecisaoLinear = precision_score(Y_Teste, LinearPredict)
PrecisaoPolinomial = precision_score(Y_Teste, PolinomialPredict)
PrecisaoRBF = precision_score(Y_Teste, RBFPredict)
PrecisaoSigmoid = precision_score(Y_Teste, SigmoidPredict)
# calculo da revocação
RevocacaoLinear = recall_score(Y_Teste, LinearPredict)
RevocacaoPolinomial = recall_score(Y_Teste, PolinomialPredict)
RevocacaoRBF = recall_score(Y_Teste, RBFPredict)
RevocacaoSigmoid = recall_score(Y_Teste, SigmoidPredict)
# Calculo do F1 score
F1_ScoreLinear = f1_score(Y_Teste, LinearPredict)
F1_ScorePolinomial = f1_score(Y_Teste, PolinomialPredict)
F1_ScoreRBF = f1_score(Y_Teste, RBFPredict)
F1_ScoreSigmoid = f1_score(Y_Teste, SigmoidPredict)

print("Acurácia Linear: ", AcuraciaLinear)
print("Acurácia Polinomial: ", AcuraciaPolinomial)
print("Acurácia RBF: ", AcuraciaRBF)
print("Acurácia Sigmoid: ", AcuraciaSigmoid)

print()

print("Precisão Linear: ", PrecisaoLinear)
print("Precisão Polinomial: ", PrecisaoPolinomial)
print("Precisão RBF: ", PrecisaoRBF)
print("Precisão Sigmoid: ", PrecisaoSigmoid)

print()

print("Revocação Linear: ", RevocacaoLinear)
print("Revocação Polinomial: ", RevocacaoPolinomial)
print("Revocação RBF: ", RevocacaoRBF)
print("Revocação Sigmoid: ", RevocacaoSigmoid)

print()

print("F1-Score Linear: ", F1_ScoreLinear)
print("F1-Score Polinomial: ", F1_ScorePolinomial)
print("F1-Score RBF: ", F1_ScoreRBF)
print("F1-Score Sigmoid: ", F1_ScoreSigmoid)
```

```python
import matplotlib.pyplot as plt

Matriz_de_Confusao = confusion_matrix(Y_Teste, PolinomialPredict, labels=ClassificadorSVM_Polinomial.classes_)

Display = ConfusionMatrixDisplay(confusion_matrix=Matriz_de_Confusao, display_labels=ClassificadorSVM_Polinomial.classes_)

Display.plot()

plt.show()
```

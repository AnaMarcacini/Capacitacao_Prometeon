```python
Classes = DataFrame["Species"].value_counts().index
```

```python
for i in range(0, len(DataFrame)):
    if DataFrame.at[i, "Species"] == Classes[3]:
        DataFrame.at[i, "Species"] = Classes[2]
        
    elif DataFrame.at[i, "Species"] == Classes[4]:
        DataFrame.at[i, "Species"] = Classes[1]
    
    elif DataFrame.at[i, "Species"] == Classes[5]:
        DataFrame.at[i, "Species"] = Classes[1]
    
    elif DataFrame.at[i, "Species"] == Classes[6]:
        DataFrame.at[i, "Species"] = Classes[0]

DataFrame["Species"].value_counts()
```

```python
SepalLengthCmOutliers = DataFrame[ (DataFrame["SepalLengthCm"] <= 2) | (DataFrame["SepalLengthCm"] >= 8) ]
SepalLengthCmOutliers["SepalLengthCm"].value_counts()
```

```python
DataFrame.drop(SepalLengthCmOutliers.index, inplace=True)
```

```python
SepalWidthCmOutliers = DataFrame[(DataFrame["SepalWidthCm"] <= 1.9) | (DataFrame["SepalWidthCm"] >= 5)]
SepalWidthCmOutliers["SepalWidthCm"].value_counts()
```

```python
DataFrame.drop(SepalWidthCmOutliers.index, inplace=True)
```

```python
PetalLengthCmOutliers = DataFrame[(DataFrame["PetalLengthCm"] <= 0.5) | (DataFrame["PetalLengthCm"] >= 7)]
PetalLengthCmOutliers["PetalLengthCm"].value_counts()
```

```python
DataFrame.drop(PetalLengthCmOutliers.index, inplace=True)
```

```python
PetalWidthCmOutliers = DataFrame[(DataFrame["PetalWidthCm"] <= 2) | (DataFrame["PetalWidthCm"] >= 5)]
PetalWidthCmOutliers["PetalWidthCm"].value_counts()
```

```python
DataFrame["Species"].value_counts()
```

```python
X = []
Y = []

for i in range(0, len(DataFrame)):
    Atributos = []
    
    for atributo in range(1,5):
        Atributos.append(DataFrame.iat[i,atributo])
    
    X.append(Atributos)
    Y.append(DataFrame.iat[i,5])

print("X (ENTRADA): ", X[0])
print("Y (SAÍDA): ", Y[0])
```

```python
from sklearn import svm
from sklearn.metrics import *
from sklearn.model_selection import train_test_split

AcuraciaLinear = 0
AcuraciaPolinomial = 0
AcuraciaRBF = 0
AcuraciaSigmoid = 0

Iteracoes = 1000

for i in range(0, Iteracoes):

    X_Treinamento, X_Teste, Y_Treinamento, Y_Teste = train_test_split(X, Y, test_size=0.3)

    ClassificadorSVM_Linear = svm.LinearSVC()
    ClassificadorSVM_Polinomial = svm.SVC(kernel="poly")
    ClassificadorSVM_RBF = svm.SVC(kernel="rbf")
    ClassificadorSVM_Sigmoid = svm.SVC(kernel="sigmoid")

    ClassificadorSVM_Linear.fit(X_Treinamento, Y_Treinamento)
    ClassificadorSVM_Polinomial.fit(X_Treinamento, Y_Treinamento)
    ClassificadorSVM_RBF.fit(X_Treinamento, Y_Treinamento)
    ClassificadorSVM_Sigmoid.fit(X_Treinamento, Y_Treinamento)

    LinearPredict = ClassificadorSVM_Linear.predict( X_Teste )
    PolinomialPredict = ClassificadorSVM_Polinomial.predict( X_Teste )
    RBFPredict = ClassificadorSVM_RBF.predict( X_Teste )
    SigmoidPredict = ClassificadorSVM_Sigmoid.predict( X_Teste )

    AcuraciaLinear += accuracy_score(Y_Teste, LinearPredict)
    AcuraciaPolinomial += accuracy_score(Y_Teste, PolinomialPredict)
    AcuraciaRBF += accuracy_score(Y_Teste, RBFPredict)
    AcuraciaSigmoid += accuracy_score(Y_Teste, SigmoidPredict)

AcuraciaLinear /= Iteracoes
AcuraciaPolinomial /= Iteracoes
AcuraciaRBF /= Iteracoes
AcuraciaSigmoid /= Iteracoes
```

```python
print("Acurácia Linear: ", AcuraciaLinear)
print("Acurácia Polinomial: ", AcuraciaPolinomial)
print("Acurácia RBF: ", AcuraciaRBF)
print("Acurácia Sigmoid: ", AcuraciaSigmoid)
```
